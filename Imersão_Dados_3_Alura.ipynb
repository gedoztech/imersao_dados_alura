# Aula 4 - Merge de dados e análise de resultados

```python
dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()

# A tabela mostra os mecanismos de ação da droga
# Que são a interação bioquímica com o alvo, 
# que pode ser proteína de uma celula ou de um patógeno por exemplo
# Este mecanismo de ação traz 2 informações (separadas por _):
# O alvo e a ação sobre esse alvo
# O alvo pode ser uma enzima
# A ação é o inibidor
# O agonista é o que está potencializando aquele receptor
# O inibidor é o que está inibindo
# Temos agonista e antagonista
```

```python
# Na tabela acima, só aparecem zeros, vamos analisar se há diferentes valores
# Mostrando valores únicos de uma coluna
# O resultado é 0 ou 1, que quer dizer que é o resultado dos experimentos da tabela anterior
# Quando mecanismo é ativado, valor 1
# Quando não é ativado, recebe valor 0
dados_resultados['acat_inhibitor'].unique()
```

```python
# Mostrando contagem das ativações (valor 1) ocorridas nos experimentos
# Quantas vezes cada mecanismo de ação foi ativo em todo o conjunto de dados
# Colunas
contagem_moa = dados_resultados.select_dtypes('int64').sum().sort_values(ascending=False)
contagem_moa
```

```python
# Mostrando contagem das ativações por linha
dados_resultados.drop('id', axis=1).sum(axis=1)
# No resultado, a segunda linha deu 0
# Nesses que deu 0, temos que saber se é controle
# Ou às vezes uma droga que não ativou nada
# Pergunta: os controles ativaram algum mecanismo de ação?
# Esperamos que não ative nada
```

```python
# Para responder a questão acima, temos que trabalhar com nossas 2 bases de dados
# dados e dados_resultados
# Base de dados com os experimentos e base de dados com os resultados
# Precisamos fazer um merge entre as tabelas

# Primeiro, criar uma nova coluna na tabela para armazenar com os totais de ativos (1)
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
# Depois, criar nova coluna na tabela para armazenar informação se cada linha teve ativo ou não
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)
dados_resultados.head()
```

```python
# Depois:
# Combinando as 2 tabelas
# Trazendo todas as colunas da tabela dados
# Trazendo apenas as colunas que interessa da tabela dados_resultados para esta situação
# Utilizando a coluna id como chave primária
dados_combinados = pd.merge(dados, dados_resultados[['id', 'n_moa', 'ativo_moa']], on='id')
dados_combinados.head()
```

```python
# Finalmente, respondendo a pergunta:
# Mostrando os totais para a coluna ativo_moa da tabela combinada
# Apenas para os experimentos de controle
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
# Resultado como o esperado, nenhum controle foi ativado por algum mecanismo de ação
```

```python
# Outra pergunta:
# Mostrado se houve alguma substância que não ativou o mecanismo de ação
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
# O resultado mostra os totais mas não é possível a interpretação sem saber por droga
# Pois o alvo pode ser de uma bactéria, ou de um vírus que está no nosso corpo
# Então pode ser interessante que não ative as nossas células humanas
# Mas que haja, tenha algum efeito no patógeno
# Então, no caso, depende da droga
```

```python
# Gerando gráfico par analisar o gene g0 em relação aos 5 compostos mais frequentes
composto_principal = dados_combinados['composto'].value_counts().index[:6]
plt.figure(figsize=(12, 8))
sns.boxplot(data = dados_combinados.query('composto  in @composto_principal'), y='g0', x = 'composto', hue='ativo_moa')
# Resultado:
# Para uma análise mais apurada, é necessário considerar outras variáveis como dose e tempo
# Necessário quebrar cada boxplot nestas outras variáveis
# Pois este gráfico está mostrando o comportamento de um conjunto de dados
# E se separarmos conforme dose e tempo, por exemplo, 
# Teremos uma melhor visão do que acontece em cada diferente combinação de variáveis
```
