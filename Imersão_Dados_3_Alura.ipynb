# Aula 3 - Correlações, causalidade e relações entre genes

```python
# Mostrando cruzamento de dados entre as colunas dose e tempo
pd.crosstab(dados['dose'], dados['tempo'])
```

```python
# Mostrando cruzamento de dados entre as colunas tratamento, dose e tempo
# As colunas se referem à classe tratamento e as linhas às classes dose e tempo
pd.crosstab([dados['dose'], dados['tempo']], dados['tratamento'])
```

```python
# Mostrando cruzamento de dados entre as colunas tratamento, dose e tempo
# Agora de forma normalizada, considerando a soma de cada coluna igual a 1
pd.crosstab([dados['dose'], dados['tempo']], dados['tratamento'], normalize=True)
```

```python
# Mostrando cruzamento de dados entre as colunas tratamento, dose e tempo
# De forma normalizada, considerando a soma de cada linha igual a 1
# Agora é possível comparar os dados de uma coluna com outra
pd.crosstab([dados['dose'], dados['tempo']], dados['tratamento'], 
            normalize='index')
```

```python
# Comparando um outro valor (g0, por exemplo) e sendo este valor a média
pd.crosstab([dados['dose'], dados['tempo']], dados['tratamento'], 
            values=dados['g0'], aggfunc='mean')
```

```python
# Gerando gráfico de dispersão para as variáveis g0 e g3
# Cada ponto é uma linha da tabela
# ponto (g0, g3)
sns.scatterplot(x='g0', y='g3', data=dados[['g0', 'g3']])
```

```python
# Gerando gráfico de dispersão para as variáveis g0 e g8
# Cada ponto é uma linha da tabela
# ponto (g0, g8)
# resultado indica uma relação inversa
sns.scatterplot(x='g0', y='g8', data=dados[['g0', 'g8']])
```

```python
# Gerando gráfico de dispersão
# resultado indica uma relação inversa
# quando um aumenta, o outro diminui
sns.lmplot(data=dados, x='g0', y='g8', line_kws={'color':'red'})
```

```python
# Gerando gráfico de dispersão
# Agora separando por tipos de tratamento
# no resultado, parece que a droga está surgindo efeito
# e o controle está "ok"
sns.lmplot(data=dados, x='g0', y='g8', line_kws={'color':'red'}, 
           col='tratamento')
```

```python
# Gerando gráfico de dispersão
# Agora separando por tipos de tratamentoe por tipos de tempo
# no resultado, parece que com mais tempo o efeito é um pouco maior
# parece que a curva vai ficando inclinada
# e o controle segue o mesmo, que é como se espera que seja
sns.lmplot(data=dados, x='g0', y='g8', line_kws={'color':'red'}, 
           col='tratamento', row='tempo')
```

```python
# Calculando a correção entre todas as colunas que vão de g0 a g711
# Quanto mais próximo de 1 ou -1, mais forte a relação entre elas
# Quanto mais próximo de 0, menor a força de relação
# Quando valor é positivo, a relação é direta
# Quando o valor é negativo, a relação é inversa
# Mas correlação não quer dizer necessariamente uma relação de causalidade
dados.loc[:, 'g0':'g771'].corr()
```

```python
# Gerando matriz de correlação para as colunas de g0 a g50
# mapa de calor entre as correlações
import numpy as np
corr = dados.loc[:, 'g0':'g50'].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Gerando matriz de correlação para as colunas de c0 a c50
# mapa de calor entre as correlações
import numpy as np
corr = dados.loc[:, 'c0':'c50'].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
