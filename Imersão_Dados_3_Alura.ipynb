# Aula 2 - Estatísticas, visualizações de dados e distribuições

```python
# renomeando a coluna droga para composto
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
# mostrando 5 primeiras linhas para conferir a mudança da coluna
dados.head()
```

```python
# Mostrando os nomes dos 5 compostos com maior frequência
cod_compostos = dados['composto'].value_counts().index[:5]
cod_compostos
```

```python
# Mostrando os dados da tabela filtrados pelos 5 compostos com maior frequência
dados_compostos = dados.query('composto in @cod_compostos')
dados_compostos
```

```python
#importando biblioteca
import seaborn as sns

# carregando configurações padrões do seaborn
sns.set()

# aumentando o tamnaho da imagem através do matplotlib
plt.figure(figsize=(8, 6))

# criando gráfico utilizando o seaborn
ax = sns.countplot(x='composto', data=dados_compostos)

# Configurando título e mostrando o gráfico
ax.set_title('Top 5 Compostos')

# Removendo texto de fora da figura
plt.show()
```

```python
# Mostrando histograma da coluna g0 com alteração da escala dos valores
dados['g0'].hist(bins=100)
```

```python
# Mostrando histograma da coluna g19 com alteração da escala dos valores
# Para comparar com g0
dados['g19'].hist(bins=100)
```

```python
# Mostrando a média dos dados da coluna g0
dados['g0'].mean()
```

```python
# Mostrando dados estatísticos utilizando comando do pandas
dados.describe()
```

```python
# Mostrando dados estatísticos das colunas que vão de g0 a g771 (todos os genes)
dados.loc[:, 'g0':'g771'].describe()
```

```python
# Alternando colunas e linhas
dados.loc[:, 'g0':'g771'].describe().T
```

```python
# Mostrando histograma de todas as médias de todos os genes (g0 a g771)
dados.loc[:, 'g0':'g771'].describe().T['mean'].hist(bins=30)
```

```python
# Mostrando histograma de todas as médias de todos os genes (c0 a c99)
dados.loc[:, 'c0':'c99'].describe().T['mean'].hist(bins=50)
```

```python
# Plotando o gráfico boxplot utilizando seaborn
sns.boxplot(x='g0', data=dados)
```

```python
# Plotando boxplot das colunas g0 e tratamento para compará-las
plt.figure(figsize=(10,8))
sns.boxplot(x='g0', y='tratamento', data=dados)
```

Acredito que, quando o gráfico está distorcido, ele possui uma resolução muito baixa para os elementos contidos nele e fica difícil a visualização clara das informações, tanto visuais, quanto de textos, principalmente se temos uma fonte muito pequena ou elementos pequenos e próximos entre si.
```python
# Desafio 1 => ordernar o gráfico countplot
ax = sns.countplot(x='composto', data=dados_compostos, 
                   order = dados_compostos['composto'].value_counts().index)
```

```python
# Desafio 2 => Melhorando a visualização do gráfico
sns.set(font_scale = 1.5)
plt.figure(figsize=(10, 8))
ax = sns.countplot(x='composto', data=dados_compostos, 
                   order = dados_compostos['composto'].value_counts().index)
ax.set_title('Top 5 Compostos')
plt.xlabel('Composto', fontsize=18)
plt.ylabel('Qtde de Registros', fontsize=18)
plt.show()
```
