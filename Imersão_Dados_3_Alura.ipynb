```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# lendo arquivo remoto através de URL e mostrando os dados
dados_url = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(dados_url, compression = 'zip') 
dados
```

```python
# mostrando valores únicos da coluna tratamento
dados['tratamento'].unique()
```

```python
# mostrando totais de registros para cada valor único da coluna tratamento
dados['tratamento'].value_counts()
```

```python
# mostrando as 5 primeiras linhas do arquivo
dados.head()
```

```python
# mostrando totais de registros de forma normalizada
# para cada valor único da coluna dose
dados['dose'].value_counts(normalize = True)
```

```python
# desafio 2 => mostrando as 5 últimas linhas do arquivo
dados.tail()
```

```python
# desafio 3 => mostrando proporção da coluna tratamento
dados['tratamento'].value_counts(normalize = True)
```

```python
# desafio 4 => mostrando total de tipos de droga utilizados
len(dados['droga'].unique())
```

```python
# desafio 5 => utilizando método query
dados.query('`g-0` > 0')
```

```python
# desafio 6 => removendo os hífens dos nomes das colunas
dados.columns = dados.columns.str.replace('-', '')
dados.columns
```

```python
# desafio 7 => gráficos com título e labels
plt.figure(figsize=(5,5))
plt.pie(dados['tempo'].value_counts(), explode=(0.01, 0.01, 0.01), labels=[72, 48, 24], autopct='%1.1f%%')
plt.title('Tempo de Ação da Droga')
plt.show()
```

```python
# cruzando dados das colunas tratamento e droga
# tentando responder Desafio 1
dados.groupby(['tratamento', 'droga']).size()
```

```python
 # cruzando dados das colunas tratamento, dose e tempo
 # tentando responder Desafio 1
 dados.groupby(['tratamento', 'dose', 'tempo']).size()
```

```python
 # cuzando dados das colunas tratamento, droga, dose e tempo
 # tentando responder Desafio 1
 dados.groupby(['tratamento', 'droga', 'dose', 'tempo']).size()
```
